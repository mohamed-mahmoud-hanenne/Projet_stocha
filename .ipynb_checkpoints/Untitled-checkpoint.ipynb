```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Charger et nettoyer le fichier CSV
filename = 'test.csv'  # Remplacez par le chemin réel du fichier
data = pd.read_csv(filename)

# Convertir les colonnes nécessaires en float
data['Cours'] = data['Cours'].astype(float)
data['Tendance'] = data['Tendance'].astype(float)

# Paramètres Q-learning
alpha = 0.1  # Taux d'apprentissage
gamma = 0.9  # Facteur de discount
epsilon = 0.1  # Probabilité d'explorer une action aléatoire
num_episodes = 1000  # Nombre d'épisodes d'apprentissage

# États et actions possibles
states = np.sort(data['Cours'].unique())  # États basés sur les cours
actions = ['Buy', 'Sell', 'Hold']
Q = np.zeros((len(states), len(actions)))  # Initialisation de la table Q

# Fonction pour obtenir l'index de l'état courant
def get_state(cours):
    return np.where(states == cours)[0][0]

# Listes pour stocker les récompenses cumulées par action
rewards_buy = []
rewards_sell = []
rewards_hold = []

# Algorithme Q-learning avec ε-greedy
for episode in range(num_episodes):
    state = get_state(data['Cours'].iloc[0])  # État initial
    total_rewards = [0, 0, 0]  # Stocke les récompenses pour Buy, Sell, Hold

    for i in range(1, len(data)):
        next_state = get_state(data['Cours'].iloc[i])
        reward = data['Tendance'].iloc[i]

        # Choix de l'action selon ε-greedy
        if np.random.rand() < epsilon:
            action = np.random.choice([0, 1, 2])  # Exploration
        else:
            action = np.argmax(Q[state, :])  # Exploitation (choix de la meilleure action)

        # Mise à jour de la table Q
        Q[state, action] = Q[state, action] + alpha * (reward + gamma * np.max(Q[next_state, :]) - Q[state, action])

        # Stocker les récompenses par action
        total_rewards[action] += reward

        # Passer à l'état suivant
        state = next_state

    # Stocker la somme des récompenses cumulées par action pour l'épisode
    rewards_buy.append(total_rewards[0])
    rewards_sell.append(total_rewards[1])
    rewards_hold.append(total_rewards[2])

# Tracer la courbe des récompenses cumulées pour chaque action
plt.figure(figsize=(10, 5))
plt.plot(np.cumsum(rewards_buy), label='Buy', linestyle='-', marker='o')
plt.plot(np.cumsum(rewards_sell), label='Sell', linestyle='--', marker='s')
plt.plot(np.cumsum(rewards_hold), label='Hold', linestyle='-.', marker='^')

# Personnalisation du graphique
plt.xlabel('Épisodes')
plt.ylabel('Récompense cumulée')
plt.title('Comparaison des stratégies: Buy, Sell, Hold')
plt.legend()
plt.grid(True)

# Affichage du graphique
plt.show()

# Affichage de la table Q finale
print("Table Q finale :")
print(Q)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Génération de données simulées
np.random.seed(42)
data_size = 500
cours = np.linspace(1, 100, data_size)  # Simulation des taux de change
tendance = np.random.choice([-1, 0, 1], size=data_size)  # Simulation des tendances (-1: baisse, 0: stable, 1: hausse)

data = pd.DataFrame({'Cours': cours, 'Tendance': tendance})

# Paramètres Q-learning
alpha = 0.1  # Taux d'apprentissage
gamma = 0.9  # Facteur de discount
epsilon = 0.1  # Exploration
num_episodes = 1000  # Nombre d'épisodes
actions = ['Buy', 'Sell', 'Hold']
states = np.sort(data['Cours'].unique())  # États basés sur les valeurs uniques des cours
Q_table = np.zeros((len(states), len(actions)))  # Initialisation de la table Q

# Fonction pour obtenir l'index d'un état
def get_state(cours):
    return np.where(states == cours)[0][0]
    
# Entraînement de SARSA
def train_sarsa():
    Q_sarsa = np.zeros((len(states), len(actions)))
    rewards_sarsa = []

    for episode in range(num_episodes):
        state = get_state(data['Cours'].iloc[0])
        action = np.random.choice([0, 1, 2]) if np.random.rand() < epsilon else np.argmax(Q_sarsa[state, :])
        total_reward = 0

        for i in range(1, len(data)):
            next_state = get_state(data['Cours'].iloc[i])
            reward = data['Tendance'].iloc[i]

            # Sélection de la prochaine action selon la politique ε-greedy
            next_action = np.random.choice([0, 1, 2]) if np.random.rand() < epsilon else np.argmax(Q_sarsa[next_state, :])

            # Mise à jour de la table Q selon SARSA
            Q_sarsa[state, action] += alpha * (reward + gamma * Q_sarsa[next_state, next_action] - Q_sarsa[state, action])

            state, action = next_state, next_action
            total_reward += reward

        rewards_sarsa.append(total_reward)

    return rewards_sarsa, Q_sarsa

# Exécution de SARSA
rewards_sarsa, Q_sarsa = train_sarsa()

# Tracer les résultats de SARSA
plt.figure(figsize=(10, 5))
plt.plot(np.cumsum(rewards_sarsa), label='SARSA', linestyle='-', marker='o', color='orange')
plt.xlabel('Épisodes')
plt.ylabel('Récompense cumulée')
plt.title('Performance de SARSA')
plt.legend()
plt.grid(True)
plt.show()
```
